```python
import pymysql
import pandas as pd
```

```python
# MySQL 연결 정보 설정
db_host = '127.0.0.1'
db_user = 'car_accident'
db_password = '1111'
db_name = 'accident'
```

```python
# MySQL 연결 생성
connection = pymysql.connect(
    host=db_host,
    port=3306, 
    user=db_user, 
    password=db_password, 
    db=db_name,
    charset='utf8mb4'
)

print(type(connection))
```

```python
df2 = pd.read_csv("./서울시_일별_시간별_교통사고_현황_utf.csv")
df2
```

```python
# MySQL Cursor 생성
cursor = connection.cursor()

# 테이블 생성 쿼리
create_table_query = '''
    CREATE TABLE IF NOT EXISTS accident_data (
        발생일 DATE NOT NULL,
        발생시간 TIME,
        발생지_시군구 VARCHAR(255) NOT NULL,
        사고건수 INT
    )
'''

# 테이블 생성
cursor.execute(create_table_query)

# UTF-8로 인코딩된 CSV 파일을 읽기
df_utf8 = pd.read_csv("서울시_일별_시간별_교통사고_현황_utf.csv")

# 데이터프레임을 MySQL 테이블로 전송
for index, row in df_utf8.iterrows():
    insert_query = f'''
        INSERT INTO accident_data 
        (발생일, 발생시간, 발생지_시군구, 사고건수) 
        VALUES 
        ('{row['발생일']}', '{row['발생시간']}', '{row['발생지_시군구']}', {row['사고건수']});
    '''
    cursor.execute(insert_query)

# 변경사항 커밋
connection.commit()
```

```python
# 연결 닫기
connection.close()
```
